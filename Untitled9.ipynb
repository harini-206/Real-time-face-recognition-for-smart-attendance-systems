```python
# ‚úÖ Import Libraries
from sklearn.datasets import fetch_lfw_people
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np
from deepface import DeepFace
import random
```

```python
# ‚úÖ Load Dataset (Labeled Faces in the Wild)
lfw_people = fetch_lfw_people(min_faces_per_person=20, resize=0.4)
print("‚úÖ Dataset Loaded")
```

```python
# ‚úÖ Display info
print("Number of images:", len(lfw_people.images))
print("Image shape:", lfw_people.images[0].shape)
print("Number of unique people:", len(lfw_people.target_names))

# ‚úÖ Create a folder to store known faces
os.makedirs("faces", exist_ok=True)
```

```python
# ‚úÖ Save one image per person in 'faces/' folder (limit to 30)
unique_people = np.unique(lfw_people.target)[:30]  # limit to first 30 people
for person_id in unique_people:
    person_name = lfw_people.target_names[person_id]
    image = lfw_people.images[np.where(lfw_people.target == person_id)][0]
    image_path = f"faces/{person_name.replace(' ', '_')}.jpg"
    cv2.imwrite(image_path, image)

print("‚úÖ Saved 30 known faces in 'faces/' folder")
```

```python
# ‚úÖ Initialize Attendance (Blank)
attendance = {name.replace("_", " "): "" for name in lfw_people.target_names[:30]}
print("\nüìù Attendance Sheet Initialized (Blank):")
print(attendance)
```

```python
# ‚úÖ Simulate Attendance Checking for Random Images
print("\nüéØ Starting Face Recognition...\n")

for i in range(10):  # randomly test with 10 images
    idx = random.randint(0, len(lfw_people.images)-1)
    test_image = lfw_people.images[idx]
    test_person = lfw_people.target_names[lfw_people.target[idx]]

    # Save test image temporarily
    cv2.imwrite("test_face.jpg", test_image)

    # DeepFace Recognition
    result = DeepFace.find(img_path="test_face.jpg", db_path="faces/", enforce_detection=False)

    if len(result[0]) > 0:
        # Extract recognized person's name
        recognized_name = result[0]['identity'][0].split("/")[-1].split(".")[0].replace("_", " ")
        attendance[recognized_name] = "Present"
        print(f"‚úÖ Recognized: {recognized_name}")
    else:
        print("‚ùå No match found")
        # Display the test image with actual and predicted names
    plt.imshow(test_image, cmap='gray')
    plt.title(f"Actual: {test_person}")
    plt.axis('off')
    plt.show()
```

```python
# ‚úÖ Mark remaining people as Absent if still blank
for person in attendance:
    if attendance[person] == "":
        attendance[person] = "Absent"

# ‚úÖ Final Attendance Sheet
print("\nüìã Final Attendance Sheet:")
for person, status in attendance.items():
    print(f"{person}: {status}")

# ‚úÖ Total Present Count
total_present = sum(1 for status in attendance.values() if status == "Present")
print(f"\nTotal Present: {total_present}/{len(attendance)}")
```
